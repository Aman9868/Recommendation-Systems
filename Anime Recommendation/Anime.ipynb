```python
import numpy as np
import pandas as pd
```

READING DATA

```python
anime=pd.read_csv('D:/Recommendation System/Anime/anime.csv')
ratings=pd.read_csv('D:/Recommendation System/Anime/rating.csv')
anime.head()
```

MERGING DATA

```python
df = pd.merge(ratings,anime,on='anime_id')
df.head()
```

```python
# CHECK UNIQUE VALUES
df.apply(lambda x : len(x.unique()))
```

```python
df.groupby('name')['rating_y'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('genre')['rating_x'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('name')['rating_y'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('name')['rating_y'].count())
ratings.head()
```

```python
moviemat = df.pivot_table(index='user_id',columns='name',values='rating_y')
moviemat.head()
```

# TOP 10 NAME RATING#

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
deathnote_user_ratings = moviemat['Death Note']
naruto_user_ratings = moviemat['Naruto']
naruto_user_ratings.head()
```

We can then use corrwith() method to get correlations between two pandas series:

```python
similar_to_deathnote = moviemat.corrwith(deathnote_user_ratings)
similar_to_naruto = moviemat.corrwith(naruto_user_ratings)
corr_deathnote = pd.DataFrame(similar_to_deathnote,columns=['Correlation'])
corr_deathnote.dropna(inplace=True)
corr_deathnote.head()
```

Now if we sort the dataframe by correlation, we should get the most similar movies, however note that we get some results that don't really make sense. This is because there are a lot of movies only watched once by users who also watched star wars (it was the most popular movie).

```python
corr_deathnote.sort_values('Correlation',ascending=False).head(10)
```


Let's fix this by filtering out movies that have less than 100 reviews (this value was chosen based off the histogram from earlier).

```python
corr_deathnote = corr_deathnote.join(ratings['num of ratings'])
corr_deathnote.head()
```

Now sort the values and notice how the titles make a lot more sense:

```python
corr_deathnote[corr_deathnote['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

Now the same for the Naruto:

```python
corr_naruto = pd.DataFrame(similar_to_naruto,columns=['Correlation'])
corr_naruto.dropna(inplace=True)
corr_naruto = corr_naruto.join(ratings['num of ratings'])
corr_naruto[corr_naruto['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
